# Plots

```python
import pandas as pd
df = pd.read_csv('decryption_results.csv')

# The second column contains the time taken values, but its name is '0.01759' due to how the CSV was saved.
# Let's get the column name for the time values dynamically.

time_col = df.columns[1]

plt.figure(figsize=(10, 5))
plt.plot(df.index, df[time_col], marker='o', linestyle='-', label='Time Taken')
plt.xlabel('Sample Index')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Taken to Decrypt Each Message')
plt.legend()
plt.grid(True)
plt.show()
```
